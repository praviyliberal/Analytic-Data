```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Change_ZP (Region, year, path_file): # функция которая берет из файла среднюю зарплату за каждый год по заданному региону 
                                         # Region - название области/республики, year - год с которого начать статистику, path_file - путь к файлу 
    
    n = 2024 - year
    Chage_df = pd.DataFrame(columns=['Год', 'Средняя зарплата']) # новый датафрейм который будет выводиться 
    new_df = pd.DataFrame(columns=['Год', 'Средняя зарплата']) # датафрейм, который будут записываться данные в цикле 
    
    df = pd.read_excel(path_file, sheet_name= str(year)) # читается файл 
    
    for i in range (n+1):
        
        df = pd.read_excel(path_file, sheet_name= str(year))
        df.rename(columns={
        "Столбец1": "Name",
    }, inplace=True)
        new_df['Год'] = year
        new_df['Средняя зарплата'] = df.loc[df['Name'] == Region]['средняя зарплата']
    
        Chage_df = pd.concat([new_df, Chage_df]) #перезапись значения в датафрейм 
        year = year + 1
    Chage_df = Chage_df.fillna(year-n-1)
    return Chage_df
```

```python
def Rename_df (name_df):
    name_df.rename(columns={
        "Столбец1": "Name",
        "среднее значение": "СЗ",
        "средняя зарплата": "СреЗ",
        "январь": "jan",
        "февраль": "feb",
        "февраль/январь": "feb/jan",
        "март": "march",
        "март/февраль": "march/feb",
        "апрель": "apr",
        "апрель/март": "apr/march",
        "май": "may",
        "май/апрель": "may/apr",
        "июнь": "june",
        "июнь/май": "june/may",
        "июль": "jule",
        "июль/июнь": "jule/june",
        "август": "aug",
        "август/июль": "aug/jul",
        "сентябрь": "sep",
        "сентябрь/август": "sep/aug",
        "октябрь": "oct",
        "октябрь/сентябрь": "oct/sep",
        "ноябрь": "nov",
        "ноябрь/октябрь": "nov/oct",
        "декабрь": "dec",
        "декабрь/ноябрь": "dec/nov",
    }, inplace=True)
```

```python
def GraphicPlot (ylName, xlName, yName ,xName, df_Name): # ylName, xlName - название осей x и y
                                                     # yName ,xName - название столбцов датафрейма
                                                     # df_Name - название датафрейма
    plt.ylabel(str(ylName))
    plt.xlabel(str(xlName))
    plt.plot(df_Name[str(xName)] , df_Name[str(yName)])
```
